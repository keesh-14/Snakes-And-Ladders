```python
import random
'''The position of snakes and ladders are stored in start:end format
where start denotes the position of ladder/snake and end denotes the
value that it reaches.'''
s={17 : 7, 54 : 34, 62 : 19, 98 : 79}#position of snake [start:end]
l={3 : 38, 24 : 33, 42 : 93, 72 : 84}#position of ladder [start:end]
```

```python
#Function to check if the entered input is either Alphanumeric or a string or an integer 
def check(s):
    return s.isdecimal()
```

```python
print("##### Welcome to Snakes and Ladders Game #####")
p1=input("Enter Name of Player1 ")
p2=input("Enter Name of Player2 ")
print("##### Let us Start #####")
flag=0 # If flag is 0 that means it's player 1's turn and if flag is 1 that means it's player 2's turn
x=0    # x Stores the position of player 1
y=0    # y Stores the position of player 2
while True:#Here we run an infinite loop
    if flag==0: 
        dice=input("Player 1 ")#dice will store the input which will be given by the user
        '''if the user inputs a string like roll,quit or any random string than this 
        check function will return false and it will enter the if condition'''
        if(not check(dice)):
            dice=dice.lower()#lower() function will convert the enterd string by the user to lower case.
            if(dice=="roll"):#if the input string is roll then random.randint() function will give generate a number between 1-6
                dice=random.randint(1,6)
            elif dice=="quit":#if player 1 quits player 2 wins.
                print("Player 2 won the Game")
                break
            else:          #if the input string is neither "roll" nor "quit" then we are printing Invalid! 
                print("Invalid!!")
                continue
        else:
            '''if the user inputs a numeric value and if it is not between 1-20(inclusive) 
    then we print Invalid and ask the user to again give a value between 1-20(inclusive).'''
       
            dice=int(dice)
            if(dice<1 or dice>20):
                print("Invalid!! Please enter a number Between 1-20 or Roll again")
                continue
        flag=1
        print("You got a",dice)
        x=x+dice           #player 1's position is updated
        if x>100:          #to prevent overshoot condition
            x=x-dice
            print("Your Final Position is :",x)
            continue   
        if x in s.keys():  #if snake encountered position is updated according to dictionary
            x =s[x]
        elif x in l.keys():#if ladder encountered position is updated according to dictionary
            x=l[x]
        print("Your Final Position is",x)
        if(x==100):        #winning condition
            print("Player 1 won the Game")
            break
        flag=1             #to alternate between turns
    
    ######## PLAYER 2 TURN ########  
    else:
        dice=input("Player 2 ")
        if(not check(dice)):
            dice=dice.lower()
            if(dice=="roll"):
                dice=random.randint(1,6)
            elif dice=="quit":
                print("Player 1 won the Game")
                break
            else: 
                print("Invalid!")
                continue
        else:
            dice=int(dice)
            if(dice<1 or dice>20):
                print("Invalid!! Please enter a number Between 1-20 or Roll again")
                continue
        print("You got a",dice)
        flag=0
        y=y+dice
        if y>100:
            y=y-dice
            print("Your Final Position is :",y)
            continue
        if y in s.keys():
            y =s[y]
        elif y in l.keys():
            y=l[y]
        print("Your Final Position is",y)
        if(y==100):
            print("Player 2 won the Game")
            break
print("##### Game Succesfully Finished #####")
```
